#**Detección de emociones**

Entrenaremos un sistema de clasificación de emociones usando un modelo de red neuronal custom. El modelo luego se usará para la detección de emociones en video en tiempo real, usando webcam.

## **Dataset:**

El dataset usado se denomina "Face expression recognition dataset" y se encuentra público en kaggle. Puedes encontrarlo en el siguiente enlace:

[Dataset Kaggle](https://www.kaggle.com/jonathanoheix/face-expression-recognition-dataset)

## **Tipos de emociones en la clasificación:**

*   angry
*   disgust
*   fear
*   happy
*   neutral
*   sad
*   surprise


### Prueba del modelo:

```python
import cv2
cap = cv2.VideoCapture(0)

faceClassif = cv2.CascadeClassifier('C:/Users/erick/Downloads/web/Face_Emotion-master/face_detector/haarcascade_frontalface2.xml')
```

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout, BatchNormalization, Input, AveragePooling2D,Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import load_model
import cv2 
import numpy as np
from tensorflow.keras.preprocessing.image import img_to_array
import matplotlib.pyplot as plt

model = load_model("modelFEC.h5")
class_names = ['angry','disgust','fear','happy','neutral','sad','surprise']

while True:
    ret,frame = cap.read()
    if ret == False: break
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    auxFrame = gray.copy()

    faces = faceClassif.detectMultiScale(gray,1.3,5)
    
    # Procesar cada cara detectada
    for (x, y, w, h) in faces:
        rostro = gray[y:y+h, x:x+w]
        rostro = cv2.resize(rostro, (48, 48))
        face2 = img_to_array(rostro)
        face2 = np.expand_dims(face2, axis=0)
        
        # Hacer la predicción de la emoción
        preds = model.predict(face2)
        
        # Mostrar el resultado
        emotion = class_names[np.argmax(preds)]
        cv2.putText(frame, emotion, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 0), 2, cv2.LINE_AA)
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
    
    # Mostrar el frame con las predicciones
    cv2.imshow('Emotion Recognition', frame)
    
    # Salir del bucle al presionar 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Liberar recursos
cap.release()
cv2.destroyAllWindows()
```
